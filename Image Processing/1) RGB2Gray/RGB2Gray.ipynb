#RGB to Gray Scale image

```python
import matplotlib.pyplot as plt
import matplotlib.image as img
from PIL import Image
import cv2
import numpy as np
```

```python
arr = cv2.imread('/content/Eiffel.jpg')
arr = cv2.cvtColor(arr, cv2.COLOR_BGR2RGB) # convert simple image to 3 channel RGB format
arr.shape
```

```python
plt.imshow(arr)
plt.title('my image')
plt.show()
```

```python
red = arr.copy()
red= arr.copy()
red[:, :, 1]= 0
red[:, :, 2]= 0

green= arr.copy()
green[:, :, 0]= 0
green[:, :, 2]= 0

blue= arr.copy()
blue[:, :, 0]= 0
blue[:, :, 1]= 0

gray= np.dot(arr, [0.299, 0.587, 0.114]) 
"""
dot product of Array and the luminance of gray RGB,
which can be modified easily by setting different numbers
"""
```

```python
fig, ax = plt.subplots(ncols = 3, figsize= (20, 8))

ax[0].set_title('Red Channel')
ax[0].imshow(red)

ax[1].set_title('Green Channel')
ax[1].imshow(green)

ax[2].set_title('Blue Channel')
ax[2].imshow(blue)

plt.show()
```

```python
plt.title('Gray Scale image')
plt.imshow(gray, cmap="gray")
plt.show()
```

```python
gray
```

```python
img.imsave('/content/Eiffel_gray.jpg', gray, cmap="gray")
```
